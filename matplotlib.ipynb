# MatPlotlib
Matplotlib is a Python library that allows you to create static, animated, and interactive visualizations. Matplotlib is a cross-platform data visualization and graphical plotting package for Python and its numerical extension NumPy. As such, it provides an open source alternative to MATLAB. Developers can also utilize matplotlib's APIs (Application Programming Interfaces) to embed charts in GUI programs.It is of the popular plotting library for python. It allows to control every aspects of a plot. 
Here is the link of the matplotlib official website: [matplotlib](http://https://matplotlib.org/)

```python
import matplotlib.pyplot as plt
```

%matplotlib inline   #to show plot within jupyter notebook
#or add plt.show() in other than anything in jupyter notebook

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

# Functional Method

```python
#functional method for plot
plt.plot(x,y,'r-')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
```

## Creating Multiplot

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Object-Oriented Method
### We have a lot more control on the figure comparing with functional approach.

```python
fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title of the plot')
```

```python
fig=plt.figure()    #it will create an empty canvus figure
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger plot')

axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

## Sub plots


```python
fig, axes = plt.subplots(nrows=2, ncols=3)   #play around changing the number of rows and column
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2)
for cu_ax in axes:
    cu_ax.plot(x,y,'r')
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2) #arranging plots in rows and column

axes[0].plot(y,x,'r')
axes[0].set_title("The red line")

axes[1].plot(x,y,'g')
axes[1].set_title("The green line")
```

## Figure Size and DPI
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig=plt.figure(figsize=(3,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(figsize=(3,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

# Saving Figure

# Legends, labels and titles

```python
#Legend

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend(loc=3)
```

```python
# Lots of options for legend locations

ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner
ax.legend(loc=10) # center
ax.legend(loc=0) #let matplotlib to choose location





# Commonly used
ax.legend(loc=0) # let matplotlib decide the optimal location
fig
```
